```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
prov_silver = pd.read_csv(os.path.join('data', 'Provident_silver.csv'))
prov_silver = prov_silver.set_index('Unnamed: 0')
prov_silver
```

```python
plt.figure(figsize=(8,5))
plt.plot(prov_silver)
plt.legend(labels=prov_silver.columns, bbox_to_anchor=(1, 1))
plt.xticks(np.arange(len(prov_silver), step=30), 
           [dt.replace('_', '/') for dt in prov_silver.index[np.arange(len(prov_silver), step=30)]], rotation=45)
plt.title('Provident Metals - Silver', size=22)
plt.ylabel('Dollars ($)', size=18);
```

```python
def makePlot(vendor, metal):
    df = pd.read_csv(os.path.join('data', '{}_{}.csv'.format(vendor, metal)))
    df = df.set_index('Unnamed: 0')
    
    plt.figure(figsize=(8,5))
    ax = plt.plot(df)
    plt.legend(labels=df.columns, bbox_to_anchor=(1, 1))
    plt.xticks(np.arange(len(df), step=30), 
               [dt.replace('_', '/') for dt in df.index[np.arange(len(df), step=30)]], rotation=45)
    if vendor == 'Provident':
        title = 'Provident Metals'
    else:
        title = 'JM Bullion'
    
    plt.title(title + " " + metal.title(), size=22)
    plt.ylabel('Dollars ($)', size=18)
    return ax
```

```python
makePlot('Provident', 'silver');
```

```python
makePlot('JM', 'platinum');
```

```python
makePlot('Provident', 'gold');
```
