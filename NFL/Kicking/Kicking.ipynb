```python
# returns one year
kicking.read_year_to_combine(1968, season='reg', ranges=False).sort_values(by='Player')
```

```python
# combines stats in range(2010, 2020) from season type 'post'
total_df = kicking.combine_years('post', start=1966, to=2020, teams=False)
```

```python
sums = total_df.drop(['Lng'], axis=1).sum() # might not have teams
num_kickers = len(total_df)
fg_acc = sums['FGM'] / sums['FG Att']
xp_acc = sums['XPM'] / sums['XPA']
print("Total fg accuracy: ", 100 * fg_acc,"%")
print("Total xp accuracy: ", 100 * xp_acc,"%")
sums
```

```python
total_df.sort_values(by='FGM', ascending=False)[:10]
```

```python
total_df['FGAA'] = total_df['FGM'] - fg_acc * total_df['FG Att']
total_df['XPAA'] = total_df['XPM'] - xp_acc * total_df['XPA']
total_df #= total_df.drop(['Team'], axis=1)
```

```python
total_df.sort_values(by='FGAA', ascending=False)
```

```python
acc_50 = sums['50+_M'] / sums['50+_A']
total_df['50+AA'] = total_df['50+_M'] - acc_50 * total_df['50+_A']
total_df['XPAA'] = total_df['XPM'] - xp_acc * total_df['XPA']
total_df.sort_values(by='50+AA', ascending=False)
```

```python
# avg acc over the years - plot
acc_dict = {}
for year in range(1966, 2020):
    sums = kicking.read_year_to_combine(year, season='reg', ranges=False).sum()
    acc_dict[year] = sums['FGM'] / sums['FG Att']
```

```python
acc_series = pd.Series(acc_dict)
acc_series.plot()
```
